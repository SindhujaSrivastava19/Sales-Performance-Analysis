# **Business Objective**
* A company wants to analyze the last year's sales data.
* Idenfying the top performing trends.
* Understanding seasonal trends and profitability patterns.

# **Import Libraries and Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Uploading our dataset

data = pd.read_csv("sales_data_2024.csv")
```

```python
# Overview of dataset

data.head()
```

```python
# Properties of dataset. Find out datatypes. If the datatypes are incorrect then fix it.

data.info()
```

```python
# In the previous step we can see that 'OrderDate' datatype should not be in object dtype. It should be in datatime dtype.

data['OrderDate'] = pd.to_datetime(data['OrderDate'])
```

```python
# Find out properties of daatype again.

data.info()
```

```python
# Data exploration. For in depth knowledge of dataset

data.describe()
```

```python
# Create column. We want 'YearMonth' column. Here there should be no data mentioned.

data['YearMonth'] = data['OrderDate'].dt.to_period('M')
```

```python
# Check the new column

data.head()
```

# **Data Analysis and Visualization**

```python
# Monthly Sales Performance

monthly_sales = data.groupby('YearMonth')['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)

plt.figure(figsize = (12, 6))
sns.lineplot(x = 'YearMonth', y= 'Sales', data = monthly_sales, marker = 'o')
plt.title("Monthly Sales Trends in 2024")
plt.ylabel("Total Sales")
plt.show()
```

**Plot Insights**
* August has the highest performance.
* March as the lowest performance.

```python
# Sales by Region

region_sales = data.groupby('Region')['Sales'].sum().reset_index().sort_values(by = 'Sales', ascending = False)

plt.figure(figsize = (8, 4))
sns.barplot(x="Region", y="Sales", data=region_sales, hue="Region", palette="viridis", legend=False)
plt.title("Total Sles by Region")
plt.ylabel("Sales")
plt.show()
```

The low sales in West region due to market standards.

```python
# Top Products

product_sales = data.groupby("Product")['Sales'].sum().reset_index().sort_values(by = 'Sales', ascending = False).head(10)
plt.figure(figsize = (10, 6))
sns.barplot(x = 'Sales', y = 'Product', data = product_sales, hue = 'Product', palette = 'mako', legend = False)
plt.xlabel("TotalSales")
plt.show()
```

```python
# Profit v/s Sales

plt.figure(figsize = (10, 6))
sns.scatterplot(x = "Sales", y = "Profit", data = data, hue = "Region", palette = "deep", legend = False)
plt.title("Profit vs Sales by Region")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

Even though the sales are increasing, there is loss.

```python
# Category Wise Sales

category_sales = data.groupby("Category")['Sales'].sum().reset_index()
fig = px.pie(category_sales,names = "Category", values = "Sales", title = "Category-wise Sales Distribution")
fig.show()
```

Electronics category hold the major portion.

But Accessories category is more than 30%, so it cannot be ignored.

# **Recommendations for the Company**


**Key Findings**
* East and North regions generate the highest revenue.
* Top Products: Tablets and Monitors.
* Sales peak in May and August.
Dips in March and July.
* No direct correlation between high sales and high profit.

**Business Insights and Impact**
* Elecronics are the main sales driver. Keep them as a priority.
* Seasonal dips need further investigation. Plan promotions to counter low months.
* West Region's performance can be boosted with targeted marketing.
* Review pricing and discount policies to avoid negative profit sales.
